```python
mode_type = df['channeltype'].mode()[0]
```

```python
print(mode_type)
```

```python
df['channeltype'] = df['channeltype'].fillna(mode_type)
```

```python
df.info()
```

```python
Top_1000_Channels=df.head(1000)
```

```python
Top_1000_Channels
```

```python
def Channel_type_distribution(Function):
    return Function[['channeltype']].value_counts()
```

```python
Channel_wise_distribution=Channel_type_distribution(Top_1000_Channels)
```

```python
print(Channel_wise_distribution)
```
